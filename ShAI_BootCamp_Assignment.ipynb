```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load your dataset
df = pd.read_csv('/content/Salaries.csv')
```

**Basic Data Exploration**

```python
num_rows, num_columns = df.shape
print("Number of rows:", num_rows)
print("Number of columns:", num_columns)
print("\ndata type valu::e\n")
print(df.dtypes)
print("\nmissing value::\n")
df.isna().sum()
```

**Descriptive Statistics:**

```python

print("mean of salary: ",df['TotalPay'].mean().round(2))
print("median of salary: ",df['TotalPay'].median())
print("mode of salary: ",df['TotalPay'].mode())
print("min of salary: ",df['TotalPay'].min())
print("max of salary: ",df['TotalPay'].max())
print("range of salary: [",df['TotalPay'].min(), " - ",df['TotalPay'].max(),']')
print("std of salary: ",df['TotalPay'].std())
```

**Data Cleaning:**

```python
df.isna().sum()
```

```python
df.drop(['Notes','Benefits','Status','Agency'],axis=1,inplace=True)
df= df.fillna(df.mean())
df.dropna(inplace=True)
df.isna().sum()
'''These columns were canceled because they did not contain data, or even a lot of them,
in addition to the Agency column because it contained one value.'''
```

**Basic Data Visualization**

```python
plt.hist(df['TotalPay'])
```

```python
dep = df['JobTitle'].value_counts()
plt.pie(dep)
plt.show()
```

**Grouped Analysis**:

```python
groubyData =df.groupby(df['Year'])
```

```python
groubyData['TotalPay'].plot(legend=True)
```

```python
groubyData['TotalPay'].describe()
```

**Simple Correlation Analysis:**

```python
sns.heatmap(df.corr())
```

```python
df.corr()
```
